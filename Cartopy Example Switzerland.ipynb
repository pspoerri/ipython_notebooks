```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (24,24)

import cartopy
import matplotlib.pyplot as plt
```

```python
borders = cartopy.feature.NaturalEarthFeature(
    category='cultural',
    name='admin_0_countries_lakes',
    scale='10m',
    edgecolor='black',
    facecolor='none')

# Data source: http://www.naturalearthdata.com/downloads/10m-physical-vectors/

# Inspiration for setup: http://scitools.org.uk/cartopy/docs/v0.13/_modules/cartopy/feature.html (see bottom)
# http://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/physical/ne_10m_rivers_lake_centerlines_scale_rank.zip
rivers = cartopy.feature.NaturalEarthFeature(
    'physical', 
    'rivers_lake_centerlines_scale_rank', # rivers_lake_centerlines_scale_rank 
    '10m',
    edgecolor=cartopy.feature.COLORS['water'],
    facecolor='none')

# Source: http://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/physical/ne_10m_rivers_europe.zip
rivers_eu = cartopy.feature.NaturalEarthFeature(
    'physical', 
    'rivers_europe', 
    '10m',
    edgecolor=cartopy.feature.COLORS['water'],
    facecolor='none')

# http://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/physical/ne_10m_lakes.zip
lakes = cartopy.feature.NaturalEarthFeature(
    'physical', 
    'lakes', 
    '10m',
    edgecolor='face',
    facecolor=cartopy.feature.COLORS['water'])

# http://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/physical/ne_10m_lakes_europe.zip
lakes_eu = cartopy.feature.NaturalEarthFeature(
    'physical', 
    'lakes_europe', 
    '10m',
    edgecolor='face',
    facecolor=cartopy.feature.COLORS['water'])
```

```python
ax = plt.axes(projection=cartopy.crs.PlateCarree())
ax.add_feature(borders, edgecolor='black')
ax.add_feature(rivers)
ax.add_feature(rivers_eu)
ax.add_feature(lakes)
ax.add_feature(lakes_eu)
ax.set_extent([5.5, 11, 45.5 ,48])
plt.show()
```

```python
ax = plt.axes(projection=cartopy.crs.PlateCarree())
# For backgrounds see: http://scitools.org.uk/cartopy/docs/latest/matplotlib/geoaxes.html?highlight=background_img#cartopy.mpl.geoaxes.GeoAxes.background_img
# Example file: https://github.com/SciTools/cartopy/blob/master/lib/cartopy/data/raster/natural_earth/images.json
# Set environment file like this: os.environ["CARTOPY_USER_BACKGROUNDS"] = "/home/magik/PYTHON/cartopy/BG/"
ax.background_img(name='ne_shaded', resolution='low')
ax.add_feature(borders, edgecolor='black')
ax.add_feature(rivers)
ax.add_feature(rivers_eu)
ax.add_feature(lakes)
ax.add_feature(lakes_eu)
ax.set_extent([5.5, 11, 45.5 ,48])
plt.show()
```

```python
ax = plt.axes(projection=cartopy.crs.PlateCarree())

features = cartopy.feature.GSHHSFeature(scale='full', levels=[1,2,3,4])

ax.add_feature(features, edgecolor='blue')
ax.add_feature(borders, edgecolor='black')
ax.set_extent([5.5, 11, 45.5 ,48])
plt.show()
```
